```python
# Extracting the reseller(Consulting services) data from main data
```

```python
df_filtered = df1.loc[(df1['Company Type'].str.contains('Services'))|(df1['Company Type'].str.contains('Consultants'))|(df1['Company Type'].str.contains('Consulting'))|(df1['Company Type'].str.contains('Consultant'))]
```

```python
df_filtered
```

```python
df_resellers = pd.DataFrame(df_filtered)
df_resellers.head()
```

```python
df_resellers.shape
```

```python
df_resellers.info()
```

```python
# trim white spaces

df_resellers = df_resellers.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
```

```python
df_resellers.duplicated().sum()
```

```python
df_resellers = df_resellers.drop_duplicates()
df_resellers
```

```python
df_resellers.rename({})
```
